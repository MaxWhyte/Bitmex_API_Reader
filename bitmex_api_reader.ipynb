```python
# Imports all necessary packages
import pandas as pd
import requests
import json
import time
```

```python
# input to feed the api with
date = "2021-09-03T20:00:00.002Z" # date and time, please change according to the data you need
path ="https://www.bitmex.com/api/v1/trade?symbol=.XBTUSDPI&columns=price&reverse=false&count=1000&startTime=" + date # URL to get lates 1000 ticks
```

```python
#Get latest 1000 items from API
r = requests.get(path)
```

```python
#Preparing the data
json_data = r.json()
df = pd.json_normalize(json_data)
```

```python
# Saving the file with a timestamped filename
timestr = time.strftime("%Y-%m-%d_%H-%M")
filename = "json_data_" + timestr + ".csv"
df.to_csv(filename, index=False, encoding="utf-8")
```
